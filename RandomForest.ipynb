Handling Unnecessary Column

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv', index_col=False)
```

```python
df.info()
```

```python
df.drop(columns=['id'], inplace=True)
```

```python
df
```

Handling Missing Values

```python
df.isnull().sum()
```

df['bmi'] = df.groupby(['gender', 'age'], group_keys=False)['bmi'].apply(lambda x: x.fillna(x.mean()))

```python
df = df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

Encoding Categorical Data

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output = False)
```

```python
ohetransform = ohe.fit_transform(df[['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']])
```

```python
encoded_df = pd.DataFrame(ohetransform, columns=ohe.get_feature_names_out
                          (['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']))
```

```python
df = pd.concat([df.drop(['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'],axis =1),
                      encoded_df], axis = 1)
```

```python
df.head()
```

```python
x = df.drop('stroke', axis = 1)
x
```

```python
y = df['stroke']
y
```

Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
x_train
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
columns_to_scale = ['age', 'avg_glucose_level', 'bmi',]
```

```python
sc = StandardScaler()
```

```python
x_train_scaled_columns = sc.fit_transform(x_train[columns_to_scale])
```

```python
x_train_scaled_df = pd.DataFrame(x_train_scaled_columns, columns=columns_to_scale, index=x_train.index)
```

```python
x_train[columns_to_scale] = x_train_scaled_df
```

```python
x_train
```

```python
x_test_scaled_columns = sc.fit_transform(x_test[columns_to_scale])
```

```python
x_test_scaled_df = pd.DataFrame(x_test_scaled_columns, columns=columns_to_scale, index=x_test.index)
```

```python
x_test[columns_to_scale] = x_test_scaled_df
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 1000, class_weight = 'balanced')
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
rf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print (confusion_matrix(y_test,y_pred))
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_res, y_train_res = smote.fit_resample(x_train, y_train)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train_res, y_train_res)
```

```python
from sklearn.metrics import recall_score


y_proba = model.predict_proba(x_test)[:, 1] 


y_pred_adjusted = (y_proba >= 0.2).astype(int)


recall = recall_score(y_test, y_pred_adjusted)
print("Recall with adjusted threshold:", recall)
```

```python
print (confusion_matrix(y_test, y_pred_adjusted))
```
